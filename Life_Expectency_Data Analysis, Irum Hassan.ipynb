This Data provides various columns and they provide the following information about the life expectancy in different countries.

Country: The name of the country.

Year: The year for which the data is provided.

Status: The status of the country, indicating whether it is classified as "Developing" or "Developed."

Population: The population of the country for the given year.

Hepatitis B: The Hepatitis B immunization coverage rate for the country.

Measles: The number of reported measles cases in the country.

Polio: The Polio immunization coverage rate for the country.

Diphtheria: The Diphtheria immunization coverage rate for the country.

HIV/AIDS: The prevalence rate of HIV/AIDS in the country.

Infant Deaths: The number of deaths of infants under one year of age.

Under-Five Deaths: The number of deaths of children under the age of five.

Total Expenditure: The total expenditure on healthcare as a percentage of GDP.

GDP: The Gross Domestic Product of the country.

BMI: The average Body Mass Index of the population.

Thinness 1-19 years: The prevalence of thinness among individuals aged 1-19 years.

Alcohol: The alcohol consumption per capita (in liters).

Schooling: The average number of years of schooling for the population.

Life Expectancy: The average life expectancy in years for the population.

# Dealing with Null values.
To fill the null values, I grouped the country column and calculated the mean value for the columns = ['Hepatitis B', 'Polio', 'Diphtheria','Total expenditure','GDP','BMI','thinness  1-19 years','Alcohol','Schooling']. I then wrote the code to fill the missing values in each country with the mean value of that country for the respective country. Even after filing, we still can see there are many empty cells. This is because there was some data missed for all the years in a country. to fill those values, i then calculated the mean of whole column and filled that mean_value into the empty columns.The null values after the first fill were very less relative to the orignal data. That's why we can now fill these cells with the average of complete columns as it will not impact the precision of the data much. 
I didn't entertain the population column and it still has all the null values. The population column shows some unrealeastic data variations. The flactuations among the years are unbelievable. The data is either misplaced or falsly edited here. Moreover, the data is missed for a country in all years which makes it even more difficult for us to detect the correct population stats and filling the data with the mean of complete column doesn't seem a good idea.


```python
fill_columns = ['Hepatitis B', 'Polio', 'Diphtheria','Total expenditure','GDP','BMI','thinness  1-19 years',
                'Alcohol','Schooling']
mean_values = life_expectency.groupby('Country')[fill_columns].transform('mean')
life_expectency[fill_columns] = life_expectency[fill_columns].fillna(mean_values)
mean_values
```

# Finding the outliers
The outliers in the data set was found by drawing the whisker plot. This whisker plot tells us that only the column representing the GDP value has multiple outliers. I found all these outliers by calculating the percentiles and get all those outliers in a list format. We could have replaced these values by taking the mean or median of all these values in the columns, but looking at the CSV file, I found that there were not many outliers within the data of same country for 15 years. Moreover, I grouped the countries and then drew a whisker plot again which caused a reduction in the number of outliers again. I thus didn't repllace the outliers as these outliers were not affecting the data of each column. 


```python
fig, ax = plt.subplots()
ax.boxplot(life_expectency[fill_columns])
ax.set_xticklabels(fill_columns, rotation = 90)
ax.set_ylabel('Values')
ax.set_title('Whisker Plot')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(life_expectency['GDP'], vert = False)
ax.set_xlabel('Values')
ax.set_ylabel('GDP')
ax.set_title('Whisker Plot')
plt.show()
```

```python

Q1 = np.percentile(life_expectency['GDP'], 25)
Q3 = np.percentile(life_expectency['GDP'], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = []
for data in life_expectency['GDP']:
    if (data <lower_bound) or (data >upper_bound):
        outliers.append(data)
        
print("Outliers:", outliers)
```

```python
#Hypothesis 1: Developed Countries has more life expectancy by year than developing countries.
count_status = life_expectency.groupby(['Status'])
fig, ax = plt.subplots(figsize=(12, 6))
for status, group in count_status:
    ax.plot(group.index, group['Life expectancy'], label=status)
ax.set_title('Life Expectancy by Country Status')
ax.legend()
plt.show()
```

```python
#Hypothesis 6: Average life expectency decreases with the increase in HIV/AIDS %age
bins = []
for i in range(0,51,4):
    bins.append(i)
life_expectency['Interval'] = pd.cut(life_expectency['HIV/AIDS'], bins=bins)
grouped_HIV = life_expectency.groupby('Interval')['Life expectancy'].mean()
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(grouped_HIV.index.astype(str),grouped_HIV)
plt.xticks(rotation=45)
ax.set_xlabel('HIV/AIDS Interval')
ax.set_ylabel('Average Life Expectancy')
ax.set_title('HIV/AIDS Trend with Respect to Life Expectancy')
plt.show()
```

# Evaluating the columns on the basis of years

The data also flactuate on the basis of years. In the graphs given below, I visualized the trend showing the life expectancy, deaths by diseases, and immunization against various diseases on the basis of year. 

```python
y = life_expectency.groupby('Year')['Life expectancy'].mean()
fig,ax = plt.subplots(figsize = (10,6))
ax.plot(y.index,y)
ax.set_xlabel('Year')
ax.set_ylabel('Mean Life Expectancy')
ax.set_title('Trend of Mean life expectancy by year')
plt.show()
```
